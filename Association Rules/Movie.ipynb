```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
movies_data=pd.read_csv('my_movies.csv')
movies_data
```

```python
movies_data.dtypes
```

```python
movies_data.isna().sum()
```

```python
movies_data1 = movies_data.drop(['V1','V2','V3','V4','V5'],axis=1)
movies_data1
```

```python
movies_data1.shape
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
frequent_itemset = apriori(df=movies_data1,min_support=0.05,max_len=3,use_colnames=True)
frequent_itemset
```

```python
rules = association_rules(frequent_itemset, metric="lift", min_threshold=1)
rules.sort_values('lift',ascending = False,inplace=True)
rules
```

```python
sns.scatterplot(x=rules['support'],y=rules['confidence'], data=rules)
plt.title("Association Rule")
plt.show()
```

```python
sns.boxplot(rules['lift'], data=rules)
plt.show()
```

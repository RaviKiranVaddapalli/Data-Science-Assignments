```python
x = airline_data.drop(labels='ID#',axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x1 = scaler.fit_transform(x)
```

```python
x1
```

### Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
```

```python
linkage = sch.linkage(x1,method='ward')
plt.figure(figsize=(15,10))
dendrogram = sch.dendrogram(linkage)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc=AgglomerativeClustering(n_clusters=3)
```

```python
y_hc = hc.fit_predict(x1)
```

```python
Clusters = pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
Clusters
```

```python
airline_data['Clusters'] = Clusters
```

```python
airline_data.head()
```

```python
airline_data.sort_values('Clusters')
```

```python
clus_analys = airline_data.groupby('Clusters').mean()
```

```python
clus_analys
```

```python
plt.figure(figsize=(5,5))  
plt.scatter(x1[:,0], x1[:,1], c= hc.labels_)
```

# K - Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(x1)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clusters = KMeans(n_clusters=3,random_state=123)
```

```python
clusters.fit(x1)
```

```python
clusters.labels_
```

```python
airline_data['Clusters'] = clusters.labels_
```

```python
airline_data.head()
```

```python
airline_data.groupby('Clusters').agg(['mean']).reset_index()
```
